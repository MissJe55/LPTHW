```python

#Ex 8 - Formatting Strings Manually

# .format() function/method is used to turn the defined formatter variable into strings
# values are inserted into the {} placeholders and return a new string
# Takeaway: shows how one template string can be reused to create different outputs via data insertion
# Define string once, reuse many times

formatter = "{} {} {} {}"   #Defining {} as a string w/ 4 placeholders

print(formatter.format(1, 2, 3, 4))   
#inserts integers into placeholders

print(formatter.format("one", "two", "three", "four"))  
#inserts strings into placeholders

print(formatter.format(True, False, False, True)) 
#inserts boolean values into placeholders

print(formatter.format(formatter, formatter, formatter, formatter)) 
#inserts template string into each placeholder

print(formatter.format(
    "My Name Is Wonder Woman",
    "Princess of Themyscira",
    "Amazon Defender",
    "Owner of Courage"

))
```
