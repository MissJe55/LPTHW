```python
#Ex 10 - Escape Codes in Strings

# Most Commonly Used Escape Characters
#\n   # new line
#\t   # tab
#\\   # backslash
#\"   # double quote
#\'   # single quote


tabby_cat = "\tI'm tabbed in."
persian_cat = "I'm split\non a line."

#Use when you want an actual backslash in your string
backslash_cat = "I'm \\ a \\ cat."   

fat_cat = '''
I'll do a list:
\t' cat food
\t' Fishies
\t' Catnip\n\t' Grass
'''

print(tabby_cat)
print(persian_cat)
print(backslash_cat)
print(fat_cat)
```
