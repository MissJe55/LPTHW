```python
# Jessica Suarez
# ex 6


types_of_people = 10    #declaring the variable
x = f"There are {types_of_people} types of people."     #Using an F-String 

binary = "binary" #when declaring a string use quotes
do_not = "don't"  #"                                "
y= f"Those who know {binary} and those who {do_not}."   #Using an F-String to declared variables

print(x)  #print F-String 
print(y)  #print F-String 

print(f"I said: {x}")  #Print the F-String with a print f format string
print(f"I also said: '{y}'") #Print the F-String with a print f format string

hilarious = True   #Declared variable set to Boolean 
joke_evaluation = "Isn't that joke so funny?! {}"  #empty bracket is for boolean value

print(joke_evaluation.format(hilarious)) #Applies format to a string that is already created

w = "This is the left side of..." 
e = "a string with a right side."

print(w+e)

```
